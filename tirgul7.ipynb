```python
class MockFaaS:
    def __init__(self):
        self.invocations = 0
        self.db = {}

    def invoke(self, function_name, payload):
        self.invocations += 1
        if function_name == 'indexer':
            return self._indexer(payload)
        elif function_name == 'searcher':
            return self._searcher(payload)
        else:
            raise ValueError(f"Function '{function_name}' not found")

    def _indexer(self, document):
        doc_id = document.get('id')
        content = document.get('content')
        self.db[doc_id] = content
        return f"Document {doc_id} indexed successfully."

    def _searcher(self, query_payload):
        query = query_payload.get('query', '').lower()
        results = [doc_id for doc_id, content in self.db.items() if query in content.lower()]
        return results

def demonstrate_faas():
    faas = MockFaaS()

    print("1. Event-Driven Demonstration (Indexing):")
    doc1 = {"id": "doc_1", "content": "Introduction to Cloud Computing and Microservices"}
    doc2 = {"id": "doc_2", "content": "Advanced Python programming concepts"}

    print(" ", faas.invoke('indexer', doc1))
    print(" ", faas.invoke('indexer', doc2))

    print("\n2. Independent Calls Demonstration (Searching):")
    query1 = {"query": "cloud"}
    query2 = {"query": "python"}

    result1 = faas.invoke('searcher', query1)
    print(f"  Search results for '{query1['query']}': {result1}")

    result2 = faas.invoke('searcher', query2)
    print(f"  Search results for '{query2['query']}': {result2}")

    print(f"\n3. Scalability Demonstration:")
    print(f"  Total function invocations: {faas.invocations}")
    print("  In real FaaS: These would execute in parallel with automatic scaling")

if __name__ == "__main__":
    demonstrate_faas()
```
