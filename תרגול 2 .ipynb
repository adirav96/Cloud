```python
# @title Display and Update Student Details { run: "auto" }
import json

# Load data from the file
file_path = '/content/drive/MyDrive/students.json'
with open(file_path, 'r') as f:
    students = json.load(f)

# Create a list of names for selection in the form
student_names = [s['first_name'] for s in students]

# --- Form fields ---
# Fix: Ensure student_selection default and options match 'first_name' values
student_selection = "Hod" # @param ["Hod", "Adir", "Ido", "Illya"]
favorite_program_input = "kh" # @param {type:"string"}
update_button = True # @param {type:"boolean"}

# Find the selected student
selected_student = next(s for s in students if s['first_name'] == student_selection)

# Display student details
print(f"--- Student Details: {student_selection} ---")
print(f"Email: {selected_student['email']}")
print(f"Courses: {', '.join(selected_student['courses'])}")
print(f"Interesting link: {selected_student['link']}")
print(f"Current favorite program: {selected_student.get('favorite_program', 'None')}")

# Update the file if the user checked the Update button
if update_button:
    selected_student['favorite_program'] = favorite_program_input
    with open(file_path, 'w') as f:
        json.dump(students, f, indent=4)
    print("\n✅ Record updated successfully in JSON file!")
```
