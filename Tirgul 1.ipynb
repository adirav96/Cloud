```python
import ipywidgets as widgets
from IPython.display import display, HTML

# reading from the text file
students = {}

with open("student.txt", "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip() != ""]

#each student data will be saved in 4 lines
for i in range(0, len(lines), 4):
    name = lines[i]
    email = lines[i+1]
    subjects = lines[i+2]
    link = lines[i+3]

    students[name] = {
        "email": email,
        "subjects": subjects,
        "link": link
    }

# change title style
display(HTML("<h2 style='color:#4CAF50;'>Students Form</h2>"))

# Dropdown
dropdown = widgets.Dropdown(
    options=list(students.keys()),
    description='student:'
)

# students fields
email_text = widgets.Text(description="email :", disabled=True)
subjects_text = widgets.Text(description="courses :", disabled=True)
link_text = widgets.Text(description="link :", disabled=True)

# link field update
new_link_input = widgets.Text(description="the new link is :")

# button
button = widgets.Button(
    description="update link",
    button_style='success'
)

output = widgets.Output()

# change selected student
def on_change(change):
    name = change['new']
    student = students[name]

    email_text.value = student["email"]
    subjects_text.value = student["subjects"]
    link_text.value = student["link"]
    new_link_input.value = student["link"]

dropdown.observe(on_change, names='value')

# new link method
def update_link(b):
    name = dropdown.value
    new_link = new_link_input.value

    if new_link.strip() == "":
        with output:
            output.clear_output()
            print("this field cant be empty")
        return

    students[name]["link"] = new_link
    link_text.value = new_link

# updating the line of the link
    for i in range(0, len(lines), 4):
        if lines[i] == name:
            lines[i+3] = new_link
#
    with open("student.txt", "w", encoding="utf-8") as f:
        for i in range(0, len(lines), 4):
            f.write(lines[i] + "\n")
            f.write(lines[i+1] + "\n")
            f.write(lines[i+2] + "\n")
            f.write(lines[i+3] + "\n\n")

    with output:
        output.clear_output()
        print("the link has been updated")

button.on_click(update_link)

# Showing the form
display(
    dropdown,
    email_text,
    subjects_text,
    link_text,
    new_link_input,
    button,
    output
)

# Loading for the firs time
on_change({'new': list(students.keys())[0]})
```
