```python
import json
from firebase import firebase
FBconn = firebase.FirebaseApplication('https://tirgul3-72374-default-rtdb.europe-west1.firebasedatabase.app/',None)
res=FBconn.get('/myTest1/',None)

for key in res:
    print(key+":\t", res[key])
```

```python
def add_or_update_word(word):
    word = word.lower().strip()

    current_count = FBconn.get('/words', word)

    if current_count is None:
        FBconn.put('/words', word, 1)
    else:
        FBconn.put('/words', word, current_count + 1)

def add_text_analysis(text):
    words_list = text.split()
    for w in words_list:
        add_or_update_word(w)


def display_all_words():
    all_words = FBconn.get('/words', None)
    if all_words:
        print("\n--- Word Count List ---")
        output = [f"{word}: {count}" for word, count in all_words.items()]
        print(", ".join(output))
    else:
        print("\nNo words found in database.")

while True:
    print("\nCommon Words Counter Menu:")
    print("1. Add single word")
    print("2. Add text for analysis")
    print("3. Update word count (manual)")
    print("4. Delete word")
    print("5. View all words")
    print("6. Exit")

    choice = input("Select an option (1-6): ")

    if choice == '1':
        word = input("Enter word: ")
        add_or_update_word(word)
    elif choice == '2':
        text = input("Enter text for analysis: ")
        add_text_analysis(text)
    elif choice == '3':
        word = input("Enter word to update: ")
        count = int(input("Enter new count: "))
        FBconn.put('/words', word, count)
    elif choice == '4':
        word = input("Enter word to delete: ")
        FBconn.delete('/words', word)
        print(f"Word '{word}' deleted.")
    elif choice == '5':
        display_all_words()
    elif choice == '6':
        print("Goodbye!")
        break
    else:
        print("Invalid option, try again.")
```
